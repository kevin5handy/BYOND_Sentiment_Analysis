## Reviews Scraping

```python
scrapreview = reviews_all(
    'co.id.bankbsi.superapp',
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=2000
)
```

```python
import csv

with open('review_BYOND.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Review'])
    for review in scrapreview:
        writer.writerow([review['content']])
```

```python
review_BYOND_df = pd.DataFrame(scrapreview)
review_BYOND_df.shape
review_BYOND_df.head()
```

```python
review_BYOND_df.to_csv('review_BYOND.csv', index=False)
```
